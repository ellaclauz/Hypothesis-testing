Determining whether there is a statistically significant different in bill length between Adelie species and Chinstrap species

```python
Adelie = penguins[penguins["species"] == "Adelie"]
Adelie
```

```python
Chinstrap = penguins[penguins["species"] == "Chinstrap"]
Chinstrap
```

```python
sampled_Adelie = Adelie.sample(n=30, replace= True, random_state=1)
sampled_Adelie
```

```python
sampled_Chinstrap = Chinstrap.sample(n=30, replace= True, random_state=1)
sampled_Chinstrap
```

```python
# Observed absolute difference
diff = abs(sampled_Adelie["bill_length_mm"].mean() - sampled_Chinstrap["bill_length_mm"].mean())
diff
```

```python
significance_level = 0.05
```

```python
test_results = stats.ttest_ind(a = sampled_Adelie["bill_length_mm"], b = sampled_Chinstrap["bill_length_mm"], equal_var = False)
test_results
```

```python
p_value = test_results.pvalue

p_value
```

```python
if p_value < significance_level:
    print(diff, p_value, "Reject")
else:
    print(diff, p_value, "Fail to reject")
```

With the significance level of 0.05 based on the sampled data we have at hand, we are rejecting the null hypothesis that there is no significant observed difference and we are saying that it is evident it is more likely there is a significant difference in bill length on average for Adelie species and Chinstrap species
